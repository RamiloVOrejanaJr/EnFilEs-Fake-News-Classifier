#Importing Dependencies

```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, classification_report
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
```

#Importing and preprocessing the datasets

##English Dataset

```python
# importing English Dataset
en_df = pd.read_csv("Philippine Fake News Corpus.csv") #dataset containing English
                                                      #news written in the Philippines
print(en_df['Content'][0])
en_df
```

```python
en_df = en_df.dropna()
print(en_df['Label'].value_counts())
```

```python
#resetting index
en_df.reset_index(inplace=True)

#dropping all unnecesary columns, only the label and article content is kept
en_df = en_df.drop(columns=['index', 'Headline', 'Authors', 'Date',
                                   'URL', 'Brand'])

#renaming clumn names to prepare for concatenation
en_df.rename(columns = {'Content':'article', 'Label':'label'}, inplace = True)


#shuffling article order then resetting shuffled index again
en_df = en_df.sample(frac=1, random_state=42).reset_index(drop=True)

#getting only 5500 articles from the dataset
en_df = en_df[0:5500]
en_df
```

##Filipino Dataset

```python
# importing Filipino Dataset
#dataset contains Filipino news articles written in the Philippines
fil_df = pd.read_csv("fil_dataset.csv") #dataset csv file is originally named
                                        #full.csv, file was renamed here for
                                        #convenience purposes
```

```python
#converting labels of Filipino Dataset to match labels of English Dataset
#according to the Github repository of the dataset, 1 means fake and 0 means real news
fil_df = fil_df.dropna()
fil_df['label'] = fil_df['label'].apply(lambda c: 'Not Credible' if c == 1 else 'Credible')
print((fil_df['label']).value_counts())
fil_df
```

##Spanish Dataset

```python
# importing Spanish Dataset
sp_df1 = pd.read_excel("development.xlsx")
sp_df2 = pd.read_excel("test.xlsx")
sp_df3 = pd.read_excel("train.xlsx")
sp_df1
```

```python
print((sp_df1['Category']).value_counts())

#converting labels of dataset to match labels of English Dataset
sp_df1['Category'] = sp_df1['Category'].apply(lambda c: 'Not Credible' if c == 'Fake' else 'Credible')
print((sp_df1['Category']).value_counts())
sp_df1
```

```python
#Renaming columns to make identical column names to prepare for concatenation
sp_df2.rename(columns = {'ID':'Id',
                         'CATEGORY':'Category',
                         'TOPICS':'Topic',
                         'SOURCE':'Source',
                         'HEADLINE':'Headline',
                         'TEXT':'Text',
                         'LINK':'Link'}, inplace = True)

print((sp_df2['Category']).value_counts())

#converting labels of dataset to match labels of English Dataset
#print(type((sp_df2['Category'])[0])) <-- outputs boolean, so:
#the following line checks for boolean instead of string value like sp_df1 and 3
sp_df2['Category'] = sp_df2['Category'].apply(lambda c: 'Credible' if c == True
                                              else 'Not Credible')

print((sp_df2['Category']).value_counts())
sp_df2
```

```python
print((sp_df3['Category']).value_counts())

#converting labels of dataset to match labels of English Dataset
sp_df3['Category'] = sp_df3['Category'].apply(lambda c: 'Not Credible' if c == 'Fake' else 'Credible')
print((sp_df3['Category']).value_counts())
sp_df3
```

```python
sp_dataset = pd.concat([sp_df1, sp_df2, sp_df3], ignore_index=True)
print(sp_dataset['Category'].value_counts())
sp_dataset.dropna(inplace=True)
sp_dataset.reset_index(inplace=True)
print(sp_dataset['Category'].value_counts())
sp_dataset
```

```python
sp_dataset.drop(columns=['index','Id', 'Topic', 'Source', 'Headline', 'Link'], inplace=True)
sp_dataset.rename(columns = {'Category':'label', 'Text':'article'}, inplace = True)

#converting labels of Spanish Dataset to match labels other datasets
#sp_dataset['label'] = sp_dataset['label'].apply(lambda c: 'Not Credible' if c == 'Fake' else 'Credible')
print((sp_dataset['label']).value_counts())
sp_dataset
```

#Splitting Data

```python
#Splitting English Dataset
x = en_df["article"]
y = en_df["label"]
en_x_train,en_x_test,en_y_train,en_y_test = train_test_split(x, y, test_size = 0.3, random_state=50, shuffle=True)  # 30% of data reserved for testing
print(en_x_train.head())


#Splitting Filipino Dataset
x = fil_df["article"]
y = fil_df["label"]
fil_x_train, fil_x_test, fil_y_train, fil_y_test = train_test_split(x, y, test_size = 0.3, random_state=50, shuffle=True)


#Splitting Spanish Dataset
x = sp_dataset["article"]
y = sp_dataset["label"]
es_x_train, es_x_test, es_y_train, es_y_test = train_test_split(x, y, test_size = 0.3, random_state=50, shuffle=True)
print(es_x_train.tail())
```

#Combining Datasets

```python
combined_x_train = pd.concat([en_x_train, fil_x_train, es_x_train], ignore_index=True)
combined_x_test = pd.concat([en_x_test, fil_x_test, es_x_test], ignore_index=True)
combined_y_train = pd.concat([en_y_train, fil_y_train, es_y_train], ignore_index=True)
combined_y_test = pd.concat([en_y_test, fil_y_test, es_y_test], ignore_index=True)

#print(combined_x_train.head())
#print(combined_x_train.tail())

print(combined_y_train.value_counts())
print(combined_y_test.value_counts())

#combined_train
```

#Text Preprocessing

```python
import re
def remove_quotes(input_string): #removes parts of the input string that are between double quotes ("")
    input_string = re.sub('"[^"]+"', '', input_string)
    #input_string = re.sub('“[^”]+”', '', input_string)

    return input_string
#                 'Between: "[every character that is NOT a double quotation mark]+"'
#                 '"[]"(double quotes outside brackets) = everything between double quotes'
#                 '[^"](arrow up and double quote inside bracket = class; every character that is NOT a double quotation mark
#                 ' + = multiple characters
```

```python
def remove_quotes(input_string):
    input_string = re.sub('"[^"]+"', ' ', input_string)

    return input_string
```

```python
#creating another version of dataset with direct quotes removed
combined_x_train_dqr = combined_x_train.apply(lambda c: remove_quotes(c))
combined_x_test_dqr = combined_x_test.apply(lambda c: remove_quotes(c))
```

##Feature Extraction

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#feature extraction for dataset, direct quotes NOT removed

#instantiation of CountVectorizer and TfidfTransformer
count_vect = CountVectorizer()
tf_transformer = TfidfTransformer()

#x_train uses fit transform; fit will only be done on training data
combined_x_train = count_vect.fit_transform(combined_x_train)
combined_x_train = tf_transformer.fit_transform(combined_x_train)

#x_test
combined_x_test = count_vect.transform(combined_x_test)
combined_x_test = tf_transformer.transform(combined_x_test)
```

```python
#feature extraction for COMBINED dataset WITH DQR

#instantiation of CountVectorizer and TfidfTransformer
count_vect_dqr = CountVectorizer()
tf_transformer_dqr = TfidfTransformer()

#x_train uses fit transform; fit will only be done on training data
combined_x_train_dqr = count_vect_dqr.fit_transform(combined_x_train_dqr)
combined_x_train_dqr = tf_transformer_dqr.fit_transform(combined_x_train_dqr)

#x_test
combined_x_test_dqr = count_vect_dqr.transform(combined_x_test_dqr)
combined_x_test_dqr = tf_transformer_dqr.transform(combined_x_test_dqr)
```

```python
#feature extraction for one language datasets WITH DQR (For accuracy purposes only)

#creating DQR versions of one language datasets
en_x_test_dqr = en_x_test.apply(lambda c: remove_quotes(c))
fil_x_test_dqr = fil_x_test.apply(lambda c: remove_quotes(c))
es_x_test_dqr = es_x_test.apply(lambda c: remove_quotes(c))

#getting tf-idf of individual languages
en_x_test = count_vect.transform(en_x_test)
en_x_test = tf_transformer.transform(en_x_test)

fil_x_test = count_vect.transform(fil_x_test)
fil_x_test = tf_transformer.transform(fil_x_test)

es_x_test = count_vect.transform(es_x_test)
es_x_test = tf_transformer.transform(es_x_test)

#getting tf-idf of individual languages that underwent DQR
en_x_test_dqr = count_vect_dqr.transform(en_x_test_dqr)
en_x_test_dqr = tf_transformer_dqr.transform(en_x_test_dqr)

fil_x_test_dqr = count_vect_dqr.transform(fil_x_test_dqr)
fil_x_test_dqr = tf_transformer_dqr.transform(fil_x_test_dqr)

es_x_test_dqr = count_vect_dqr.transform(es_x_test_dqr)
es_x_test_dqr = tf_transformer_dqr.transform(es_x_test_dqr)
```

#Model Training

Raw Model

```python
#making 2 linearsvc models, one for with quotes
linearsvc = LinearSVC()

#training linearsvc model on training data
linearsvc.fit(combined_x_train, combined_y_train)
```

```python
noQuotesModel = LinearSVC()
noQuotesModel.fit(combined_x_train_dqr, combined_y_train)
```

#Testing Model Performance On Overall Dataset

```python
#predicting testing data split
test_predictions = linearsvc.predict(combined_x_test)
noquotes_predictions = noQuotesModel.predict(combined_x_test_dqr)
```
